```python
import pandas as pd
dataFrame1=pd.read_csv('學生成績.csv')
dataFrame1.index=dataFrame1['姓名']
dataFrame2=dataFrame1[['國文','英文','數學','地理','歷史']]
dataFrame2
```

```python
name1=dataFrame2.loc[['蔡偉成']]
name1
```

```python
name1.loc['蔡偉成'].sum()
sum=name1.loc['蔡偉成'].sum()
mean=name1.loc['蔡偉成'].mean()
name1['總分']=sum
name1['平均']=mean
name1
```

```python
sum_s=dataFrame2.sum(axis=1)
mean_s=dataFrame2.mean(axis=1)
dataFrame2['總分']=sum_s
dataFrame2['平均']=mean_s
dataFrame2
```
