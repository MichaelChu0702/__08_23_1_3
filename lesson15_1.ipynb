```python
import matplotlib.pyplot as plt
import numpy as np
index=np.array([3,4,5,6,7])#x軸位置
subject=['apple','banana','orange','tomato','guava']
value1=[5000,6000,7000,8000,9000]
value2=[3000,4000,5000,6000,7000]
value3=[4000,5000,6000,7000,8000]

figure=plt.figure(figsize=(8,5))#建立畫布
axes=figure.add_subplot(1,1,1)#圖表
bin_width=0.3#寬度
barContainer1=axes.bar(index,value1,width=bin_width,color='red')#圖表畫圖
barContainer2=axes.bar(index-0.3,value2,width=bin_width,color='blue')#圖表畫圖
barContainer3=axes.bar(index+0.3,value3,width=bin_width,color='green')#圖表畫圖
axes.set_xticks(index,labels=subject)
xmin,xmax,ymin,ymax=axes.axis()
axes.axis([xmin-0.2,xmax+0.2,ymin,ymax+500])
for rec in barContainer1.patches:
    #print(rec)
    #print(rec.xy[0])
    height=rec.get_height()#求出高
    axes.annotate(str(height),xy=(rec.xy[0],height+80))#顯示數字

for rec in barContainer2.patches:
    #print(rec)
    #print(rec.xy[0])
    height=rec.get_height()#求出高
    axes.annotate(str(height),xy=(rec.xy[0]-0.02,height+80))#顯示數字

for rec in barContainer3.patches:
    #print(rec)
    #print(rec.xy[0])
    height=rec.get_height()#求出高
    axes.annotate(str(height),xy=(rec.xy[0],height+80))#顯示數字 rec.xy[0]抓x座標


plt.show()#顯示圖

```
