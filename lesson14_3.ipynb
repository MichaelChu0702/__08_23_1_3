```python
import matplotlib.pyplot as plt

index=[3,4,5,6,7]
subject=['apple','banana','orange','tomato','guava']
value=[5000,7000,3000,4000,6000]
#為何要抓出34567


figure=plt.figure()
axes=figure.add_subplot(1,1,1)
axes.bar(index,value)#條狀圖 長條圖
axes.set_xticks(index,labels=subject)
for idx, i in enumerate (axes.get_xticks()):#idx得到01234 i得到34567
    axes.annotate(str(value[idx]),xy=(i-0.18,value[idx]+50))
#axes.annotate(str(value[0]),xy=(3-0.18,5000+50))#在藍線上顯示5000
xmin,xmax,ymin,ymax=axes.axis()
axes.axis(xmin=xmin-0.3,xmax=xmax,ymin=1000,ymax=ymax+500)
plt.show()
```
