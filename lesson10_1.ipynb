```python
import csv
with open('學生成績.csv',mode='r',encoding='utf-8',newline='') as file:
    csvReader=csv.reader(file)
    next(csvReader)
    #for row in csvReader:
       # print(row)
    students=list(csvReader)
students
```

```python
print("國文大於80分的學生有:")
sum=0
for student in students:
    
    if int (student[1])>=80:
        sum+=1
        print(f'學生:{student[0]},國文{student[1]}')
print(f"共有{sum}位學生")
    
```
