```python
import numpy as np
import pandas as pd

s1=pd.Series([4.5,7.2,-5.3,3.6],index=['d','b','a','c'])
s1
```

```python
#選取使用subscript
s1[['a','b']]
```

```python
#選取使用reindex()
s1.reindex(['a','b'])
```

```python
dataFrame1=pd.DataFrame(np.arange(9).reshape(3,3),
             index=['a','c','d'],
             columns=['台北','台中','高雄'])
dataFrame1
```

```python
#使用loc[],subscript
dataFrame1.loc[['a','c']]
```

```python
#使用reindex()
dataFrame1.reindex(index=['a','c'])
```

```python
#使用[],subscript
dataFrame1[['台北','台中']]
```

```python
#使用reindex()
dataFrame1.reindex(columns=['台北','台中'])
```

```python
#drop(),移除列或欄
dataFrame2=pd.DataFrame(np.arange(16).reshape(4,4),
             index=['台北','台中','台南','高雄'],
             columns=['one','two','three','four'])
dataFrame2
```

```python
#drop(),移除列或欄
dataFrame2.drop(index=['台北','台中'])
```

```python
dataFrame2.loc[['台南','高雄']]
```
