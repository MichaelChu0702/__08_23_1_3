```python
class  Student:
    def __init__(self,name=str,ch=int,eng=int,math=str,geo=int,his=int):
        #attribute
        self.姓名=name
        self.國文=ch
        self.英文=eng
        self.數學=math
        self.地理=geo
        self.歷史=his
    def __repr__(self)->str:
        return f'{self.姓名}的實體'
    
    #實體方法
    def sum(self)->int:
        return self.國文+self.英文+self.數學+self.地理+self.歷史
    #實體方法
    def average(self)->float:
        return round(self.sum()/5,ndigits=2)
```

```python
import csv
with open("學生成績.csv",encoding="utf-8",newline='') as file:
    dict_reader=csv.DictReader(file)
    students=[]
    for student in dict_reader:
        
        chinese=int (student['國文'])
        english=int (student['英文'])
        math=int (student['數學'])
        geography=int (student['地理'])
        history=int (student['歷史'])
        student=Student(student['姓名'],
                ch=chinese,
                eng=english,
                math=math,
                geo=geography,
                his=history)
        students.append(student)
    

students
        
```

```python
for stu in students:
    print(stu['姓名'])
    sum=stu['國文']+stu['英文']+stu['數學']+stu["地理"]+stu["歷史"]
    print(f"總分:{sum}")
    print(f"平均:{sum/5:.2f}")
    print("=============")
```

```python
for stu in students:
    print(stu.姓名)
    print(f'總分:{stu.sum()}')
    print(f'平均:{stu.average()}')
    print("=====")
```
