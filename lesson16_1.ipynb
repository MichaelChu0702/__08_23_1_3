```python
#-二維list
#50個學生 每個學生的分數

import random
students=[]
for j in range(50):
    scores=[]
    for i in range(5):
        scores.append (random.randint(50,100))
    students.append(scores)
students
```

```python
for student in students:
    total=0
    for item in student:
        total+=item
    print(total)
```

```python
import numpy as np
scores=np.random.randint(50,high=101,size=(50,5))

sums=np.sum(scores,axis=1)
for student in sums:
    print(student)

```

```python
means=np.mean(scores,axis=1)
for student in means:
    print(student)
```
