```python
import matplotlib.pyplot as plt
values=[20,30,45,10]
labels=['Nokia','Samsung','Apple','Lumia']
colors=['yellow','green','red','blue']
figure=plt.figure()
axes=figure.add_subplot(1,1,1)
axes.pie(values,labels=labels,colors=colors,shadow=True,startangle=180,autopct='%.1f%%')

plt.show()
```
