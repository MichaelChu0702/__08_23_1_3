```python
#搜尋值
import pandas as pd
dataFrame1=pd.read_csv('學生成績.csv')
dataFrame1.index=dataFrame1['姓名']
dataFrame2=dataFrame1.drop(columns='姓名')
dataFrame2.info()
```

```python
#國文成績不及格,subscript
dataFrame2[dataFrame2['國文']<60]
```

```python
#國文成績不及格,使用query
dataFrame2.query('國文<60')
```

```python
#國文成績或者英文不及格(其一不及格),subscript
dataFrame2[(dataFrame2['國文']<60)| (dataFrame2['英文']<60)]
```

```python
#國文成績和英文不及格(同時不及格),subscript
dataFrame2[(dataFrame2['國文']<60)& (dataFrame2['英文']<60)]
```

```python
#國文成績或者英文不及格(其一不及格),query()
dataFrame2.query('國文<60 or 英文<60')
```

```python
#國文成績和英文不及格(同時不及格),query()
dataFrame2.query('國文<60 and 英文<60')
```
