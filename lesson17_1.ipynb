```python
#dataFrame排序
import numpy as np
dataFrame1=pd.DataFrame(np.arange(8).reshape((2,4)),
                        index=["three","one"],
                        columns=['d','a','b','c'])
dataFrame1.sort_index(axis=1)
dataFrame1.sort_index(axis=1,ascending=False)
dataFrame1.sort_values(by='d',ascending=False)
dataFrame1.sort_values(by='one',axis=1,ascending=False)#ascending 由大到小
                        
 
```

```python
dataFrame1=pd.read_csv('學生成績.csv')
dataFrame1.index=dataFrame1['姓名']
dataFrame2=dataFrame1[['國文','英文','數學','地理','歷史']]
dataFrame2
```

```python
sum=dataFrame2.sum(axis=1)#加總要橫的 sum接收
mean=dataFrame2.mean(axis=1)
dataFrame2['總分']=sum#建立總分項目
dataFrame2['平均']=mean#建立平均項目
dataFrame2
```

```python
#全班前五名
dataFrame2.sort_values(by='總分',ascending=False).iloc[:5]#by抓項目 ascending False大到小 iloc[:5]前五名
#倒數五名
dataFrame2.sort_values(by='總分').iloc[:5]#ascending預設是True


```

```python
dataFrame2['排名']=dataFrame2['總分'].rank(ascending=False,method='min')
```

```python
dataFrame3=dataFrame2.sort_values(by='排名')
```

```python
dataFrame3.to_csv('學生成績1.csv')
```
